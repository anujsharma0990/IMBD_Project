```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# Plot a graph to show the distribution of genre in the top 250 movies 

```python
# Load the movie data into a pandas DataFrame
df = pd.read_csv('movies.csv')

# Get the count of each genre in the top 250 movies
genre_counts = df['genre'].h.value_counts()

# Plot the graph
plt.figure(figsize=(8, 8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Genres in Top 250 Movies')
plt.axis('equal')
```
